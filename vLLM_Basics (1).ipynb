```
%%bash
/content/venv/bin/python - <<'PY'
import traceback
from vllm import LLM, SamplingParams

prompts = [
    "Hello, my name is",
    "The capital of France is"
]

# More conservative LLM config for Colab Tesla T4
# - gpu_memory_utilization: cap GPU usage to 70%
# - disable heavy caching / chunked prefill to reduce memory pressure
# - set a safer cudagraph mode (piecewise) via compilation_config to reduce graph issues
# - dtype float16 for speed/memory efficiency
compilation_config = {
    "mode": None,
    "cudagraph_mode": "PIECEWISE",   # less aggressive than FULL_AND_PIECEWISE
    "cudagraph_num_of_warmups": 1
}

try:
    params = SamplingParams(temperature=1, top_p=0.95, max_tokens=48, seed=1234)
    llm = LLM(
        model="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
        dtype="float16",
        gpu_memory_utilization=0.7,      # your requested cap
        enable_prefix_caching=False,     # reduce memory pressure
        enable_chunked_prefill=False,    # avoid some aggressive prefill behavior
        compilation_config=compilation_config
    )

    outputs = llm.generate(prompts, params)
    for i, out in enumerate(outputs):
        print("\n=== Prompt", i+1, "===\n")
        print(out.outputs[0].text.strip())
except Exception:
    traceback.print_exc()
    print("\n*** LLM init/generation failed or engine crashed. Paste the traceback if so. ***")
PY
```

```
%%bash
/content/venv/bin/python - <<'PY'
import time, traceback
from vllm import LLM, SamplingParams

# --- Prompt (few-shot anchor + question) ---
few_shot = """Q: What is the capital of Germany?
A: Berlin.

Q: What is the capital of Italy?
A: Rome.

Q: What is the capital of France?
A:"""

# --- Sampling: strict greedy (deterministic) ---
params = SamplingParams(
    temperature=0.0,   # deterministic
    top_p=1.0,
    top_k=1,           # force greedy next token
    max_tokens=40,
    seed=0
)

# --- Conservative engine config for TinyLlama on T4 ---
compilation_config = {
    "mode": None,
    "cudagraph_mode": "PIECEWISE",    # safer than FULL
    "cudagraph_num_of_warmups": 1
}

try:
    t0 = time.time()
    llm = LLM(
        model="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
        dtype="float16",
        gpu_memory_utilization=0.7,    # your requested cap
        enable_chunked_prefill=True,   # re-enable for this model (avoid warning)
        enable_prefix_caching=False,   # keep off to reduce memory pressure
        max_model_len=1024,            # reduce KV cache memory (optional)
        compilation_config=compilation_config
    )
    t1 = time.time()
    out = list(llm.generate([few_shot], params))[0]
    t2 = time.time()

    print("startup (s):", round(t1-t0,2))
    print("generate (s):", round(t2-t1,2))
    print("tokens/sec approx:", round((1*params.max_tokens)/(t2-t1),1))
    print("\n=== Output ===\n")
    print(out.outputs[0].text.strip())

except Exception:
    traceback.print_exc()
    print("\n*** LLM init/generation failed or engine crashed. Paste the traceback if so. ***")
PY
```
